```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), OOP
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot()
```

```python
# Create the plot and add a label for the legend.

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Changing line properties

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Use OOP Approach
ax = plt.axes()
ax.plot(x_axis, y_axis, label = 'Boston', color = 'green', marker = 'd', linewidth = 2)
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.grid()
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0,45)
#Change Line Color to green
```

```python
#Making markers bigger
y_axis_larger=[y*3 for y in y_axis]


plt.scatter(x_axis, y_axis, s = y_axis_larger)

##can include list comprehension within scatter

plt.scatter(x_axis, y_axis, s = [y*3 for y in y_axis])
plt.show()
```

```python
#Using OOP to create bubble charts
'''Change the color of the markers to sky blue.

Change the size of the markers to 5 times each data point.

Make the color 20% transparent.

Add a black edge color to the circles.

Make the linewidth of the circles 2 points.

Add a legend for the city of Boston.

Add a title.

Switch the axis so that the Fare($) data is on the x-axis.

Add a limit for the x-axis data.

Invert the y-axis so that January is at the top.'''


fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [y*5 for y in y_axis], c = 'skyblue', alpha = .8, edgecolors= 'k', linewidths= 2, label = 'Boston', )
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
fig.gca().invert_yaxis()


```
